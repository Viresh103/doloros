```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("train.csv")
X=pd.read_csv("test.csv")
y=train["target"]
xtrain=train.drop(["target"],axis=1)
```

```python
x=xtrain.iloc[:,[1,2,18,20]]
X=X.iloc[:,[1,2,18,20]]
```

## applying model


```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=2)
x=poly_reg.fit_transform(x)
x=x[:,[ 0, 1, 2, 4, 13]]
x=pd.DataFrame(data=x,columns=[0,1,2,3,4])
newvar1=np.divide(x.iloc[:,4:5],x.iloc[:,3:4])/100
newvar1=newvar1.rename(columns={4: "newvar1"})
x=pd.concat([x,newvar1],axis=1)
x=x.drop([1],axis=1)

```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
import keras
from keras.layers import Dense
from keras.models import Sequential
```

```python
model=Sequential()
model.add(Dense(input_dim=5,output_dim=7,activation="linear",kernel_initializer='uniform'))
model.add(Dense(output_dim=7,activation="tanh",kernel_initializer='uniform'))
model.add(Dense(output_dim=7,activation="tanh",kernel_initializer='uniform'))
model.add(Dense(output_dim=1,activation="tanh",kernel_initializer='uniform'))
```
